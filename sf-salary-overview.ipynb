```python
# the below code will keep giving error and i will explain in the next line of code why is that.

sal["BasePay"].mean()

# observe the Dtype of 'BasePay'
```

```python
# 'BasePay' has the Dtype of 'object', that is why we cannot perform a mathematical calculation on that.
# we will conver that first into 'float'
sal['BasePay'] = pd.to_numeric(sal['BasePay'], errors='coerce')
```

```python
# Average BasePay?
# now the same line of code will not give us the error
sal["BasePay"].mean()
```

```python
# Maximum overtime pay?
# same for all object dtypes to perform a maths operation.
sal['OvertimePay'] = pd.to_numeric(sal['OvertimePay'], errors='coerce')
sal['OvertimePay'].max()
```

```python
# title of employee named 'JOSEPH DRISCOLL'
sal[sal['EmployeeName'] == 'JOSEPH DRISCOLL']['JobTitle']
```

```python
# How much does "JOSEPH DRICOLL" make?\
sal[sal['EmployeeName'] == 'JOSEPH DRISCOLL']['TotalPayBenefits']
```

```python
# Employee with highest salary
sal[sal['TotalPayBenefits'] == sal['TotalPayBenefits'].max()]['EmployeeName']
```

```python
# or 2nd method
sal.loc[sal['TotalPayBenefits'].idxmax()]['EmployeeName']
```

```python
#lowest paid person
sal[sal['TotalPayBenefits'] == sal['TotalPayBenefits'].min()]['EmployeeName']
```

```python
# or 2nd method
sal.loc[sal['TotalPayBenefits'].idxmin()]['EmployeeName']
```

```python
# avergae BasePay for year 2011-2014
sal.groupby('Year').mean()['BasePay']
```

```python
#how many unique job titles in the dataset
sal['JobTitle'].nunique()
# we want number and not the list i.e. nunique
```

```python
# or 2nd method
len(sal['JobTitle'].unique())
```

```python
# Top 5 Most common job titles
sal['JobTitle'].value_counts().head(5)
```

```python
# job tiles with only one occurence in 2013
sum(sal[sal['Year'] == 2013]['JobTitle'].value_counts() == 1)
```

```python
# how many people have job title with 'Chief in it'
def chief_title(title):
    if 'chief' in title.lower().split():
        return True
    else:
        return False
```

```python
# to use a python function on Pandas we use "apply method"
sum(sal['JobTitle'].apply(lambda x : chief_title(x)))
```

```python
# correlation between Job title string and salary
sal['title_len'] = sal['JobTitle'].apply(len)
```

```python
sal[['JobTitle','title_len']]
```

```python
# to find corrleration betwen len of the title and the salary
sal[['TotalPayBenefits','title_len']].corr()
```
